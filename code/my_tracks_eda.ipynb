```python
def cosine_sim(test,pool):
    main_song_list= test[['danceability','energy', 'loudness', 'speechiness', 'acousticness', 
                          'instrumentalness', 'liveness', 'valence', 'tempo']].values.tolist()
    pool_song_list= [pool[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
                           'instrumentalness', 'liveness', 'valence', 'tempo']].values.tolist()]
     
    return cosine_similarity(main_song_list, pool_song_list)
```
